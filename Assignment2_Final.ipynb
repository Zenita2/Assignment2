# Step1:
We merged all five texts into one and standardized quotation marks, as we did in the assignment1.

```python
import os
fileids = os.listdir("texts")
fileids = [("texts/" + fileid) for fileid in fileids]
fileids
```

```python
textlist = []
for fileid in fileids:
    with open (fileid, mode ="r", encoding = "utf-8") as f:
        text = f.read()
        textlist.append(text)
textstring = str(textlist)
```

```python
pipeline = [('“', '"'), ('´´', '"'), ('”', '"'), ('’’', '"')]
for old, new in pipeline:
    textstring = textstring.replace(old, new)
textstring = textstring.replace("\\n", "")
textstring = textstring.replace("\'", "")
textstring = textstring.replace("\\", "")
```

```python
textstring
```

```python
doc1 = nlp(textstring)
doc1._.coref_chains.print()
```

```python
print(doc1._.coref_chains.resolve(doc1[563]))
```

```python
for ent in doc1.ents:
    if ent.label_ in ["PERSON"]:
        print(ent.text, ent.label_)
```

```python
from spacy import displacy
```

```python
options = {"ents": ["PERSON"],
          "colors": {"PERSON": "lightsteelblue"}}

displacy.render(doc1, style="ent", options=options, jupyter=True)
```

# Extracting Quotes


```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
```

```python
matcher = Matcher(nlp.vocab)
pattern_q = [{'ORTH': '"'}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': '"'}]
matcher.add("QUOTES", [pattern_q], greedy='LONGEST')
doc2 = nlp(textstring)
matches_q = matcher(doc2)
matches_q.sort(key = lambda x: x[1])
print (len(matches_q))
for match in matches_q[:10]:
    print (match, doc[match[1]:match[2]])
```
